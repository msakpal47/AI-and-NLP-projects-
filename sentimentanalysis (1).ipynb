```
analyzer = SentimentIntensityAnalyzer()
```

```
def sentiment_vedar(text):
    over_all_polarity = analyzer.polarity_scores(text)
    if over_all_polarity['compound'] >= 0.05:
        return 'Positive'
    elif over_all_polarity['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```
